```python
#loading dataset
path = r'C:\Users\adity\Downloads\salary.csv'
df = pd.read_csv(path)
df.head()
```

# EDA AND DATA PREPROCESSING

```python
#Summary of dataset
print(df.info(),end='\n\n')

#Finding NA values in df
print(df.isna().sum())
```

```python
#Dropping NA values
df = df.dropna()

#Verify there's no NA values
print(df.isna().sum())
```

```python
#Rename columns
df.rename(columns={
    'Age':'age',
    'Gender':'gender',
    'Education Level':'edu',
    'Job Title':'job',
    'Years of Experience':'exp'
},inplace=True)
```

```python
df.columns
```

# FEATURE ENGINEERING

```python
#Encoding categorical data
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder()

X_gender = df.loc[:,"gender"].values
X_Ed = df.loc[:,"edu"].values
X_Job = df.loc[:,"job"].values

X_gender = label_encoder.fit_transform(X_gender)
X_Ed = label_encoder.fit_transform(X_Ed)
X_Job = label_encoder.fit_transform(X_Job)

print(X_gender)
print(X_Ed)
print(X_Job)

df['gender'] = X_gender
df['edu'] = X_Ed
df['job'] = X_Job

```

# TRAINING AND VALIDATION SPLIT

```python
#Split data into train and test set

X = df.iloc[:,0:-1].values
y = df.iloc[:,-1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #80-train, 20 test

print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
y_test = y_test.reshape(len(X_test),1)
y_test
```

# FEATURE SCALING -> STANDARDIZATION

```python
#Feature Scaling of dependent and independent vars
#Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
y_train = sc.fit_transform(y_train)
X_test = sc.fit_transform(X_test)
y_test = sc.fit_transform(y_test)

print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

# LINEAR REGRESSION MODEL BUILDING

```python
#Building Regression model on Training data
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
#Testing model by Making predictions on testing data.
y_pred = reg.predict(X_test)
y_pred # return Scaled predictd values of salary
```

```python
#Model accuracy
print(reg.score(X_test,y_test))
```

```python
#Converting y_pred values from scaled to orginal
inv_transform = sc.inverse_transform(y_pred)
inv_transform #Actual predicted salary from scaled y_pred feature
```

```python
#Validating diff b/w y_test & y_pred

np.concatenate((sc.inverse_transform(y_test), sc.inverse_transform(y_pred)),axis=1)

df2 = pd.DataFrame(np.concatenate((sc.inverse_transform(y_test), sc.inverse_transform(y_pred)),axis=1), columns = ['y','yhat'])
df2

```

```python
#Model Eavluation

from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
  
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)  
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True 
rmse = root_mean_squared_error(y_true=y_test,y_pred=y_pred) 
  
print("MAE:",mae) 
print("MSE:",mse) 
print("RMSE:",rmse)
```

# ANN REGRESSION

```python
#Neural Net
import tensorflow as tf
print(tf.__version__)
```

```python
#Building NN Regression

ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=32,input_dim=2, activation='relu'))
ann.add(tf.keras.layers.Dense(units=64, activation='relu'))
ann.add(tf.keras.layers.Dense(units=64, activation='relu'))
ann.add(tf.keras.layers.Dense(units=32, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='linear'))


##Compile NN --> with SGD & MSE Loss func
ann.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy','mae'])

#Training ANN
history = ann.fit(X_train,y_train, batch_size=32, epochs=100, validation_data=(X_test, y_test))
```

```python
#predicting results on Testing set
print(ann.predict(X_test))
y_pred = ann.predict(X_test) #return standardized scaled values
```
